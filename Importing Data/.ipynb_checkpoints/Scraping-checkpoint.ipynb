```python
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
s = "https://en.wikipedia.org/wiki/Lists_of_films"
```

```python
def scrape(url,element="",arg="",skip_no_arg=True):
    
    # http request url
    response = requests.get(url)
    
    # get response text
    page = response.text
    
    # parse response using bs4
    soup = BeautifulSoup(page,"html.parser")
    
    # get elements in parsed response
    if (element == "") | (element == None):
        elements = soup.find_all("p")
    else:
        elements = soup.find_all(element)
        
    # generate elements table
    element_tables = []
    for e in elements:
        if (arg != "") & (arg != None):
            try:
                e = e[arg]
                element_tables.append(str(e))
            except KeyError:
                print("arg not found in element")
                if not(skip_no_arg):
                    element_tables.append(str(e))
        else:
            element_tables.append(str(e))
    
    print(str(len(element_tables)) + " element(s) found")
    
    return element_tables
```

```python
l = scrape(url=s,element="a",arg="href",skip_no_arg=False)
```

```python
scrape(s,"a","href")
```
