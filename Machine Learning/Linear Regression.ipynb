```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso, Ridge, LinearRegression
```

```python
# specify model target and features

target_var = 'mpg'
feature_vars = [ 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
id_var = "model"

X = data[feature_vars]
y = data[target_var]
ID = data[id_var]
```

```python
# Linear Model

lin_reg = LinearRegression()
lin_reg.fit(X=X,y=y)

print(lin_reg.score(X,y))
print(lin_reg.coef_)
plot_lm(lin_reg)
```

```python
# Ridge Model

rig_reg = Ridge(alpha=1)
rig_reg.fit(X=X,y=y)

print(rig_reg.score(X,y))
print(rig_reg.coef_)
plot_lm(rig_reg)
```

```python
# Lasso Model

lass_reg = Lasso(alpha=0.5)
lass_reg.fit(X=X,y=y)

print(lass_reg.score(X,y))
print(lass_reg.coef_)
plot_lm(lass_reg)
```
