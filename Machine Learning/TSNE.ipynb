# TSNE

```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
data.columns
```

```python
feature_vars = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
target_var = "variety"

X = data[feature_vars]
y = data[target_var]
col_dict = dict(zip(y.unique(),[1,2,3]))
```

```python
# Create a TSNE instance: model
model = TSNE(learning_rate = 10,n_components = 3)

# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(X)
```

```python
# Select the 0th feature: xs
xs = tsne_features[:,0]

# Select the 1st feature: ys
ys = tsne_features[:,1]

# Select the 1st feature: zs
zs = tsne_features[:,2]
```

```python
# Scatter plot, coloring by variety_numbers
plt.scatter(xs,ys,c=y.map(col_dict))
plt.show()
```

```python
fig = px.scatter_3d(x=xs,y=ys,z=zs,color=y)
fig.show()
```
